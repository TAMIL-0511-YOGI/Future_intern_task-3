```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload Excel file
uploaded = files.upload()
df = pd.read_excel(list(uploaded.keys())[0])

# Drop unnecessary index column
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

# **Program: Average Scores per Criterion (Bar Chart)**

```python
# Average score per criterion
avg_scores = df.mean(numeric_only=True)

plt.figure(figsize=(10,5))
avg_scores.plot(kind='bar', color='skyblue')
plt.title("Average Scores per Criterion")
plt.ylabel("Average Rating")
plt.xticks(rotation=45, ha='right')
plt.show()
```

# **Program: Distribution of Course Recommendation (Count Plot)**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Course recommendation based on relevance', data=df, palette='viridis')
plt.title("Distribution of Course Recommendation")
plt.show()
```

# **Program: Correlation Heatmap**

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Criteria")
plt.show()
```

# **Program: Boxplots for Variability**

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df.drop(columns=['Student ID']))
plt.xticks(rotation=45, ha='right')
plt.title("Boxplots of Ratings per Criterion")
plt.show()
```

# **Program: Histogram for Each Criterion**

```python
import matplotlib.pyplot as plt

for col in df.drop(columns=['Student ID']).columns:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=10, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.xlabel("Rating")
    plt.ylabel("Frequency")
    plt.show()
```

# **Program: Radar Chart (Spider Plot)**

```python
import numpy as np

criteria = df.drop(columns=['Student ID']).columns
avg_scores = df[criteria].mean()

# Radar chart setup
angles = np.linspace(0, 2*np.pi, len(criteria), endpoint=False).tolist()
scores = avg_scores.tolist()
scores += scores[:1]  # close the loop
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
ax.plot(angles, scores, 'o-', linewidth=2, label="Average Scores")
ax.fill(angles, scores, alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(criteria, rotation=45, ha='right')
ax.set_title("Radar Chart of Average Scores")
plt.show()
```

# **Program: Violin Plots**

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=df.drop(columns=['Student ID']), inner="quartile", palette="muted")
plt.xticks(rotation=45, ha='right')
plt.title("Violin Plots of Ratings per Criterion")
plt.show()
```

# **Program: Pairplot**

```python
sns.pairplot(df.drop(columns=['Student ID']))
plt.suptitle("Pairplot of Criteria Relationships", y=1.02)
plt.show()
```

# **Program: Line Chart of Average Scores**

```python
avg_scores = df.drop(columns=['Student ID']).mean()

plt.figure(figsize=(10,5))
plt.plot(avg_scores.index, avg_scores.values, marker='o', linestyle='-', color='blue')
plt.title("Line Chart of Average Scores per Criterion")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Average Rating")
plt.show()
```

# **ALL IN ONE PROGRAM :**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from google.colab import files
from matplotlib.backends.backend_pdf import PdfPages

# Step 1: Upload Excel file
uploaded = files.upload()
df = pd.read_excel(list(uploaded.keys())[0])

# Step 2: Clean dataset
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

# Step 3: Create PDF file
pdf = PdfPages("student_feedback_report.pdf")

# --- Chart 1: Average Scores per Criterion ---
avg_scores = df.drop(columns=['Student ID']).mean()
plt.figure(figsize=(10,5))
avg_scores.plot(kind='bar', color='skyblue')
plt.title("Average Scores per Criterion")
plt.ylabel("Average Rating")
plt.xticks(rotation=45, ha='right')
pdf.savefig(); plt.close()

# --- Chart 2: Distribution of Course Recommendation ---
plt.figure(figsize=(6,4))
sns.countplot(x='Course recommendation based on relevance', data=df, palette='viridis')
plt.title("Distribution of Course Recommendation")
pdf.savefig(); plt.close()

# --- Chart 3: Correlation Heatmap ---
plt.figure(figsize=(10,6))
corr = df.drop(columns=['Student ID']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Criteria")
pdf.savefig(); plt.close()

# --- Chart 4: Boxplots ---
plt.figure(figsize=(12,6))
sns.boxplot(data=df.drop(columns=['Student ID']))
plt.xticks(rotation=45, ha='right')
plt.title("Boxplots of Ratings per Criterion")
pdf.savefig(); plt.close()

# --- Chart 5: Histograms ---
for col in df.drop(columns=['Student ID']).columns:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=10, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.xlabel("Rating")
    plt.ylabel("Frequency")
    pdf.savefig(); plt.close()

# --- Chart 6: Radar Chart ---
criteria = df.drop(columns=['Student ID']).columns
avg_scores = df[criteria].mean()
angles = np.linspace(0, 2*np.pi, len(criteria), endpoint=False).tolist()
scores = avg_scores.tolist()
scores += scores[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
ax.plot(angles, scores, 'o-', linewidth=2, label="Average Scores")
ax.fill(angles, scores, alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(criteria, rotation=45, ha='right')
ax.set_title("Radar Chart of Average Scores")
pdf.savefig(); plt.close()

# --- Chart 7: Violin Plots ---
plt.figure(figsize=(12,6))
sns.violinplot(data=df.drop(columns=['Student ID']), inner="quartile", palette="muted")
plt.xticks(rotation=45, ha='right')
plt.title("Violin Plots of Ratings per Criterion")
pdf.savefig(); plt.close()

# --- Chart 8: Pairplot ---
sns.pairplot(df.drop(columns=['Student ID']))
plt.suptitle("Pairplot of Criteria Relationships", y=1.02)
pdf.savefig(); plt.close()

# --- Chart 9: Line Chart ---
avg_scores = df.drop(columns=['Student ID']).mean()
plt.figure(figsize=(10,5))
plt.plot(avg_scores.index, avg_scores.values, marker='o', linestyle='-', color='blue')
plt.title("Line Chart of Average Scores per Criterion")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Average Rating")
pdf.savefig(); plt.close()

# Step 4: Close PDF
pdf.close()

print("âœ… PDF report 'student_feedback_report.pdf' has been created.")
```
